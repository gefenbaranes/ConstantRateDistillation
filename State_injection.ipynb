```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate error input data
error_in = np.logspace(-3, -1, 20)

# Set up the figure
plt.figure(figsize=(4.3, 3.3))

# Define a single color with different shades from a colormap
color = plt.cm.Blues(np.linspace(0.4, 0.9, 3))

# Plot data for different gate error rates
for i, p2 in enumerate([1e-3, 5e-3, 1e-2]):
    p1 = p2  # 1q gate error rate

    p_injection_MR = (3/5) * p2 + (2/3) * p1
    distillation_error = (6/5) * p2 + (4/3) * p1 + error_in
    # distillation_error_2 = 1 - ((1 - p_injection_MR) ** 2) * (1 - error_in)
    plt.plot(error_in, distillation_error, label=f"{p2*100} %", linewidth=2, color=color[i])
    # plt.plot(error_in, distillation_error_2, '--' , label=f"{p2*100} try 2%", linewidth=2, color=color[i])
    

# Set the log scale for both axes
plt.xscale('log')
plt.yscale('log')
# plt.ylim(bottom = 1e-3)


plt.legend(title="Gates error rates", loc='upper left', fontsize=11, title_fontsize = 11, bbox_to_anchor=(1, 1))
plt.legend(title="Gates error rates", loc='upper left', fontsize=10, title_fontsize = 10)

plt.xlabel("Bell pair infidelity", fontsize = 14)
plt.ylabel("Distillation input error", fontsize = 14)
# plt.grid(True, which="both", linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
plt.grid(True, linestyle='--', alpha=0.7, linewidth=1)
# plt.title("Distillation Error Rate vs Bell Pair Error", fontsize=16)
plt.tight_layout()
plt.savefig(f"Distillation_error_vs_bell_injection.pdf")

plt.show()
```
